```python
import pandas as pd
df=pd.read_csv("ML_datasets/Enjoy_sports.csv")
```

```python
def finds(df):
    pr=df[df['EnjoySport']=='yes']
    pe=pr.iloc[:, :-1].values
    hypo=["#"]*len(pe[0])
    for e in pe:
        for i in range(len(hypo)):
            if hypo[i]=='#':
                hypo[i]=e[i]
            elif hypo[i]!=e[i]:
                hypo[i]='?'
    return hypo
```

```python
print(finds(df))
```

```python
import pandas as pd
df=pd.read_csv("C:/Users/LENOVO/OneDrive/Desktop/datasets/Enjoy_sports.csv")
data=df.values.tolist()
```

```python
def cae(data):
    
    S = ["#"] * (len(data[0]) - 1)  
    G=[["?"]*(len(data[0])-1)]

    for row in data:
        atri,lab=row[:-1],row[-1]

        if lab=="yes":
            for i in range(len(S)):
                if S[i]=="#":
                    S[i]=atri[i]
                elif S[i]!=atri[i]:
                    S[i]="?"
            G=[g for g in G if all (g[i]=="?" or g[i]==S[i] for i in range(len(S)))]
        
        elif lab=="no":
            new_g=[]
            for g in G:
                for i in range(len(g)):
                   
                    if g[i]=="?":
                        new_h=g[:]
                        new_h[i]=atri[i]
                        new_g.append(new_h)
            G=new_g
    return S,G
    
sf,gf=cae(data)
print(sf)
print(gf)
```

```python
import pandas as pd
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split

from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv("iris.csv")
df.head()
```

```python
x=df.iloc[:,0:4]
y=df.select_dtypes(include=[object])
```

```python
le=preprocessing.LabelEncoder()
y=y.apply(le.fit_transform)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42,stratify=y)
```

```python
scaler=StandardScaler()
scaler.fit(x_train)
x_test=scaler.transform(x_test)
x_train=scaler.transform(x_train)
```

```python
mlp=MLPClassifier(hidden_layer_sizes=(10,10,10),max_iter=2000)
mlp.fit(x_train,y_train.values.ravel())
```

```python
y_pred=mlp.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```
